```python
import numpy as np

def johnson(data, lmb, inv):
    if inv:
        data_ = data.copy()
    else:
        data_ = data
    new_data = []
    for x in data_:
        
        if x >= 0:
            if lmb == 0:
                new_data.append(np.log(x + 1)) if not inv else new_data.append(np.exp(x)-1)
            else:
                new_data.append(((x + 1)**lmb -1)/lmb) if not inv else new_data.append((lmb*x+1)**(1/lmb) -1)

        if lmb == 2:
            new_data.append(-np.log(1-x)) if not inv else new_data.append(1-np.exp(-x))
        else:
            new_data.append((1-(1-x)**(2-lmb))/(2-lmb)) if not inv else new_data.append(1-(1-(2-lmb)*x)**(1/(2-lmb)))
    return np.array(new_data)
    
```

```python
orig_data=np.linspace(-10,10,1000)
LMB=[-1,0.5,0,2,3,4]
new_data = {}
for lmb in LMB:
    new_data[lmb] = johnson(orig_data, lmb, True)

data = {}
for lmb in new_data.keys():
    data[lmb] = johnson(new_data[lmb], lmb, False)

```
