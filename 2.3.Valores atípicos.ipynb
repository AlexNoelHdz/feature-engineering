```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
LI=load_iris()
data=LI.data
names=LI.feature_names
df=pd.DataFrame(data,columns=names)
df.head()
```

```python
Q=df.quantile([0.25,0.75])
Q
```

```python
Q.loc[0.25] # Primer cuartíl
```

```python
Q.iloc[0] # Primer cuartíl (index 0)
```

```python
iqr=Q.iloc[1]-Q.iloc[0]
iqr
```

```python
Li=Q.iloc[0]-1.5*iqr
Li
```

```python
Ls=Q.iloc[1]+1.5*iqr
Ls
```

```python
df.describe()
```

```python
df1=(df<Li)|(df>Ls)
df1
```

```python
df1.sum() # Cuántos datos atípicos tenemos por variable
```
