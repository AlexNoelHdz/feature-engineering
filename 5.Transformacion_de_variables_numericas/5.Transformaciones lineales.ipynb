```python
import pandas as pd

from sklearn.datasets import fetch_california_housing
ch=fetch_california_housing()
df=pd.DataFrame(ch.data,columns=ch.feature_names)
df.head()
```

# Estandarización

Desplaza la media $\mu$ de $X$ a cero y escala la varianza $\sigma$ a uno
$$
X'=\frac{X-\mu}{\sigma},
$$
donde $X'$ es la variable transformada con media cero y varianza uno.

```python
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
```

```python
df.describe()
```

```python
df_est=zscore(df)
df_est.describe()
```

```python
escalador=StandardScaler()
pd.DataFrame(escalador.fit_transform(df),columns=df.columns).describe()
```

```python
escalador.mean_
```

```python
escalador.scale_
```

# Escalamiento con base en la media

Desplaza la media $\mu$ de $X$ a cero y el rango a uno
$$
X'=\frac{X-\mu}{max(X)-min(X)}.
$$

```python
import numpy as np
```

```python
m=df.mean()
rng=np.ptp(df,axis=0)
m
```

```python
rng
```

```python
df_m=(df-m)/rng
df_m.describe()
```

```python
np.ptp(df_m,axis=0)
```

# Escalamiento Robusto (Por cuantiles)
Desplaza la mediana $\bar{x}$ de $X$ a cero y el rango intercuartílico a uno
$$
X'=\frac{X-\bar{x}}{Q_3-Q_1},
$$
donde $Q_1$ y $Q_3$ son los cuartiles uno y tres respectivamente.

```python
from sklearn.preprocessing import RobustScaler
```

```python
me=df.median()
Q=df.quantile([0.75,0.25])
Q
```

```python
iqr=Q.loc[0.75]-Q.loc[0.25]
iqr
```

```python
df_r=(df-me)/iqr
df_r.median()
```

```python
df_r.quantile([0.75,0.25])
```

```python
_.loc[0.75]-_.loc[0.25]
```

```python
esc_rob=RobustScaler()
df_r=pd.DataFrame(esc_rob.fit_transform(df),columns=df.columns)
df_r.median()
```

```python
df_r.quantile((0.75,0.25))
```

```python
esc_rob.center_ # Medianas de 'df'
```

```python
esc_rob.scale_ # IQR de 'df'
```

# Escalamiento de máximo absoluto
$$
X'=\frac{X}{\max{|X|}}.
$$

Se recomienda para datos centrados en cero o para variables con pocas observaciones.

```python
df_max=df/np.max(np.abs(df))
df_max.describe()
```

# Normalización de mínimo y máximo
Queremos que nuestros datos esten entre (0,1).
$$
X'=\frac{X-\min{X}}{\max{X}-\min{X}}.
$$

```python
A,B=5,10
Xp=(df-df.min())/(df.max()-df.min())
X_AB=Xp*(B-A)+A
X_AB.describe()
```

# Regresión

```python
df['MedHouseVal']=ch.target
df.head(2)
```

```python
from scipy.stats import yeojohnson
from sklearn.linear_model import LinearRegression
```

```python
lmb={}
df_YJ=df.copy()
for v in df:
  df_YJ[v],lmb[v]=yeojohnson((df[v]-df[v].mean())/df[v].std())
H=df_YJ.hist(figsize=(10,8),bins=50)
```

```python
df_YJ.boxplot(figsize=(10,7))
```

```python
from sklearn.preprocessing import RobustScaler,StandardScaler
```

```python
var_da=['MedInc','AveRooms','AveBedrms','Population','AveOccup']
df_N=df_YJ.copy()
esc_rob=RobustScaler()
df_N[var_da]=esc_rob.fit_transform(df_YJ[var_da])
df_N.describe()
```

```python
var_sinDa=['HouseAge','Latitude','Longitude']
esc_est=StandardScaler()
df_N[var_sinDa]=esc_est.fit_transform(df_YJ[var_sinDa])
df_N.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
lin_N=LinearRegression()
Xn,xn,Yn,yn=train_test_split(df_N.drop(columns='MedHouseVal'),df_N['MedHouseVal'])
lin_N.fit(Xn,Yn)
```

```python
lin_N.score(Xn,Yn),lin_N.score(xn,yn)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
# Graficamos 'Population' (4), 'AveOccup' (5), 'Longitude' (7)
x=df_N['Population']
plt.scatter(x,df_N['MedHouseVal'],s=5)
plt.plot(x,lin_N.intercept_+lin_N.coef_[4]*x,'r')
plt.grid()
```

```python
x=df_N['AveOccup']
plt.scatter(x,df_N['MedHouseVal'],s=5)
plt.plot(x,lin_N.intercept_+lin_N.coef_[5]*x,'r')
plt.grid()
```

```python
x=df_N['Longitude']
plt.scatter(x,df_N['MedHouseVal'],s=5)
plt.plot(x,lin_N.intercept_+lin_N.coef_[7]*x,'r')
plt.grid()
```

```python
# Regresión con el modelo sin transformar
X,x,Y,y=train_test_split(df.drop(columns='MedHouseVal'),df['MedHouseVal'])
lin=LinearRegression()
lin.fit(X,Y)
lin.score(X,Y),lin.score(x,y)
```
